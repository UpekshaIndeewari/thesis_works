```python
#Extract the entire text from the PDF file
import PyPDF2
from PyPDF2 import PdfReader
reader = PdfReader("Unsupervised_Learning_Algorithms_for_Keyword_Extra.pdf")
text = " "
for page in reader.pages:
   text = text+ page.extract_text()
print(text)
```

```python
#Tokenization 
#split the entire paragrapgh in to sentences and words (tokens)
from nltk.tokenize import word_tokenize, sent_tokenize
list_of_words = word_tokenize(text)

print("Words in the text:\n",list_of_words)
print("Number of words in the text:\n",len(list_of_words))
```

```python
#Convert words in to lowercase
lower_case_remain =[]
for i in list_of_words:
   x = i.lower()
   lower_case_remain.append(x)
   
print("Words in lower case:\n",lower_case_remain)
print("Number of lower case words in the text:\n",len(lower_case_remain))
```

```python
#Remove cases. Remove some words and letters which are not in English language
cleaned_texts =[]
for text in lower_case_remain:
   text = text.casefold()
   cleaned_texts.append(text)
     
print("Words without cases:\n",cleaned_texts)
print("Number of words without cases:\n",len(cleaned_texts))
```

```python
#Stop words in Enlish language
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')

print("Stop words in the English language:\n",stop_words)
print("Number of stop words in Engligh language:\n",len(stop_words))
```

```python
#Stop words remove from the text
stop_word_list = []
nostop_word_list = []

for word in cleaned_texts:
   if word in stop_words:
      stop_word_list.append(word)
   elif word not in stop_words:
      nostop_word_list.append(word)
      
print("Stop word list in the text:\n",stop_word_list)
print("Number of stop words in the text:\n",len(stop_word_list))
print("List without stop words:\n",nostop_word_list)
print("Number of words in the text without stop words:\n",len(nostop_word_list))
      
```

```python
#Remove unicode symbols from the text
import re
no_unicode = []

for word in nostop_word_list:
    cleaned_word = re.sub(r'[^\x00-\x7F]+', '', word)
    no_unicode.append(cleaned_word)
    
print("List of words without uni codes:\n",no_unicode)
print("Number of words without uni codes:\n",len(no_unicode))
```

```python
#Remove numbers and digits
no_number_list = []
number_list = []

for text in no_unicode:
    tokens = word_tokenize(text)
    contains_digit =  any(token.isdigit() or (token.count('.') == 1 and token.replace('.', '').isdigit()) for token in tokens)
    
    if not contains_digit:
        no_number_list.append(text)
    elif contains_digit:
        number_list.append(text)
        
print("Words without numbers and digits:\n",no_number_list)
print("Number of words without numbers and digits:\n",len(no_number_list))
print("List of numbers and digits\n",number_list)
print("Number of digits and numbers\n",len(number_list))
```

```python
#Remove special words with having mix of string, numbers, punctuations and other symbols
import re
no_sp_words = []
num_string_list = []

# Regular expression pattern to match any string containing numeric characters or hyphens
pattern = r'^(?=.*\d)(?=.*[A-Za-z])|(?=.*[A-Za-z])(?=.*[\W_])|(?=.*\d)(?=.*[\W_])'

for text in no_number_list:
    if not re.match(pattern, text):
        no_sp_words.append(text)
    elif re.match(pattern, text):
        num_string_list.append(text)

print("Filtered list:\n",no_sp_words)
print("Number of words in the filterd list:\n",len(no_sp_words))
print("Special words in the text:\n",num_string_list)
print("Number of special words in the list:\n",len(num_string_list))
```

```python
#remove words with repetition numbers, punctuations, words and symbols
import re
cleaned_tokens = []
prev_word = None

for token in no_sp_words:
    # Remove word repetitions
    if token != prev_word or token.isnumeric() or not re.match(r'^[A-Za-z]*$', token) or token.isdigit():
        cleaned_tokens.append(token)
        prev_word = token
    else:
        continue

# Remove  repetitions
cleaned_text = ' '.join(cleaned_tokens)
cleaned_text = re.sub(r'([^\w\s])\1+', r'\1', cleaned_text)
cleaned_text = re.sub(r'(\d)\1+', r'\1', cleaned_text)
cleaned_words = cleaned_text.split()


print("List without repetitions: \n",cleaned_words)
print("The nuber of words without repetitions: \n",len(cleaned_words))
```

```python
#Remove punctuations from the list
nltk.download('punkt')
punctuation_marks=[]
without_punct_marks=[]

for token in cleaned_words:
   if token in ("!\"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"):
      punctuation_marks.append(token)
   elif token not in ("!\"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"):
      without_punct_marks.append(token)
      
print("Punctuationmarks in the list:\n",punctuation_marks)
print("Number of punctuation marks in the text:\n",len(punctuation_marks))
print("Word list without punctuation marks:\n",without_punct_marks)
print("Number of words without punctuation marks:\n",len(without_punct_marks))
```

```python
#Calculation of the percentage
Remain_percentage = (len(without_punct_marks)/len(list_of_words))*100

print("Total number of words before text cleaning is",len(list_of_words))
print("total number of words after text cleaning is",len(without_punct_marks))
print("The remaining percentage of the words is",round(Remain_percentage,2),"%")
```

```python
#Plot a graph
import matplotlib.pyplot as plt
Step = ['Total words', 'R_Lowercase', 'R_Cases', 'R_Stopwords', 'R_Unicodes', 'R_Numbers', 'R_Special words','R_rpeat','R_Punctuations']
a = len(list_of_words)
b = len(lower_case_remain)
c = len(cleaned_texts)
d = len(nostop_word_list)
e = len(no_unicode)
f = len(no_number_list)
g = len(no_sp_words)
h = len(cleaned_words)
i = len(without_punct_marks)
Number_of_words =[a,b,c,d,e,f,g,h,i]

# Find the indices of the maximum and minimum values
max_index = Number_of_words.index(max(Number_of_words))
min_index = Number_of_words.index(min(Number_of_words))

# Define colors for the bars
bar_colors = []
for i in range(len(Step)):
    if (i == max_index):
         bar_colors.append('blue')
    elif (i == min_index ):
         bar_colors.append('red')
    else:
         bar_colors.append('grey')

# figure size
plt.figure(figsize=(6, 6))

# Create the bar chart with highlighted bars
bars = plt.bar(Step, Number_of_words, color=bar_colors)
plt.xticks(rotation=90)
plt.title('Distribution of total number of words in the text during text cleaning')
plt.xlabel('Step in text cleaning')
plt.ylabel('Number of words')

# Adding labels to the bars
for i in range(len(Step)):
    plt.text(Step[i], Number_of_words[i], str(Number_of_words[i]), ha='center', va='bottom', color='black')

# Highlight the maximum and minimum bars with a different color
bars[max_index].set_color('blue')
bars[min_index].set_color('red')

plt.tight_layout()
plt.show()

```
